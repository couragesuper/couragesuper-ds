```python
def getYTNMainArticle( wd ) :
    class_ids =  ["YTN_CSA_topnews1", "YTN_CSA_topnews2", "YTN_CSA_topnews3"]
    list_mainArticle =[]
    for i in range(0,len(class_ids)) :
        elem = wd.find_element_by_class_name( class_ids[i] )
        list_mainArticle.append( elem.find_element_by_tag_name('a').get_attribute('href') )
    return list_mainArticle
        
listMainNews = getYTNMainArticle( wd )
```

```python
for i, link in enumerate( listMainNews ):
    craw.OpenWebPage(wd, link, 1)
    print( "Article %d %s" %  (i, wd.find_element_by_class_name("article_paragraph").text) )    
```
