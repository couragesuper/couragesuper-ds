```python
nSucceed = 0 
nBoard = 8 
nSelected = 0 
listClosed = [-1 for i in range(nBoard)] 
listOpen  = [False for i in range(nBoard)] 

nComputeDepth = 0 

def checkIlegal( listClosed, nSelected ) : 
    global nComputeDepth 
    nComputeDepth = nComputeDepth + 1 
    for i in range(nSelected): 
        x = i; 
        y = listClosed[i] 
        for j in range(i+1,nSelected) : 
            x2 = j 
            y2 = listClosed[j] 
            dx = abs(i - j) 
            dy = abs(y - y2) 
            if( dx == dy ) : return False 
    return True 

def solveQueen( listClosed, listOpen, nSelected, nBoard  ) : 
    global nSucceed 
    if( checkIlegal(listClosed , nSelected ) == False ) : return False 
    elif (nSelected == nBoard ) : 
        print( "nSucceed{0}".format(nSucceed),listClosed ) 
        nSucceed = nSucceed + 1 
        return True 

    for i in range( nBoard ) : 
            if( listOpen[i] == False ): 
                listOpen[i] = True 
                listClosed[nSelected] = i; 
                ret = solveQueen( listClosed, listOpen, nSelected + 1 , nBoard ) 
                #if( ret == False ) : 
                listOpen[i] = False 
                listClosed[nSelected] = -1 
    return False 
```

```python

solveQueen( listClosed, listOpen, nSelected, nBoard ) 

print( "compute depth=" , nComputeDepth )
```
